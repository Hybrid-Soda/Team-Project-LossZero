```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.yaml")  # build a new model from YAML
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo11n.yaml").load("yolo11n.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(
    data="./defect_zero/data.yaml",
    device=1,
    epochs=150,                      # í•™ìŠµ ë°˜ë³µ íšŸìˆ˜ (Epochs)
    imgsz=640,                       # ìž…ë ¥ ì´ë¯¸ì§€ í¬ê¸° (ì´ë¯¸ì§€ì˜ ê°€ë¡œ/ì„¸ë¡œ í”½ì…€ ìˆ˜)
    cache=False,                     # ë°ì´í„°ì…‹ì„ ë©”ëª¨ë¦¬ì— ìºì‹± ì—¬ë¶€ (ìºì‹± ì‹œ í•™ìŠµ ì†ë„ ì¦ê°€)
    workers=8,                       # ë°ì´í„° ë¡œë“œ ì‹œ ì‚¬ìš©í•  CPU ì›Œì»¤ ìˆ˜ (ë³‘ë ¬ ì²˜ë¦¬)
    lr0=0.01,                        # ì´ˆê¸° í•™ìŠµë¥  (Learning Rate)
    lrf=0.01,                        # í•™ìŠµë¥  ê°ì†Œ ë¹„ìœ¨ (Learning Rate Final)
    momentum=0.937,                  # ëª¨ë©˜í…€ ê°’ (Momentum)
    weight_decay=0.0005,             # ê°€ì¤‘ì¹˜ ê°ì†Œ (Weight Decay, ê³¼ì í•© ë°©ì§€)
    warmup_epochs=3.0,               # ì›Œë°ì—… Epoch ìˆ˜ (ì´ˆê¸° í•™ìŠµ ì†ë„ ì¦ê°€)
    warmup_momentum=0.8,             # ì›Œë°ì—… ë‹¨ê³„ì—ì„œ ëª¨ë©˜í…€ ê°’
    warmup_bias_lr=0.1,              # ì›Œë°ì—… ë‹¨ê³„ì—ì„œ ë°”ì´ì–´ìŠ¤ í•™ìŠµë¥  (Bias Learning Rate)
    
    # ì†ì‹¤ í•¨ìˆ˜ ê°€ì¤‘ì¹˜ ì„¤ì •
    box=7.5,                         # ë°”ìš´ë”© ë°•ìŠ¤ ì†ì‹¤ ê°€ì¤‘ì¹˜ (Box Loss)
    cls=0.5,                         # í´ëž˜ìŠ¤ ì†ì‹¤ ê°€ì¤‘ì¹˜ (Class Loss)
    dfl=1.5,                         # DFL(Distribution Focal Loss) ê°€ì¤‘ì¹˜
    pose=12.0,                       # í¬ì¦ˆ ì¶”ì • ì†ì‹¤ ê°€ì¤‘ì¹˜ (Pose Loss)
    kobj=1.0,                        # í‚¤í¬ì¸íŠ¸ ê°ì²´ ì†ì‹¤ ê°€ì¤‘ì¹˜ (Key Object Loss)

    # ë°ì´í„° ì¦ê°• ì„¤ì • (HSV ìƒ‰ ê³µê°„ ì¡°ì •)
    hsv_h=0.015,                     # ìƒ‰ì¡°(Hue) ë³€í™˜ ë²”ìœ„
    hsv_s=0.7,                       # ì±„ë„(Saturation) ë³€í™˜ ë²”ìœ„
    hsv_v=0.4,                       # ëª…ë„(Value) ë³€í™˜ ë²”ìœ„
)
```
